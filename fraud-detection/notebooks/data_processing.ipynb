```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import os

import sys
sys.path.append('../src')
#from model import 
from preprocessing import split_data, preprocess_fraud
```

```python
dd = preprocess_fraud(df)
display(dd['clean'].head(3))
display(dd['scaled'].head(3))
display(dd['pca'].head(3))
```

```python
# Save dataframes
save_paths = ['../data/clean-fraud.pkl', '../data/scaled-fraud.pkl', '../data/pca-fraud.pkl']
files = [dd['clean'], dd['scaled'], dd['pca']]
for sp, file in zip(save_paths, files):
    if os.path.isfile(sp):
        print(f'File Already Exists -- {sp}')
        pass
    else:
        pd.to_pickle(file, sp)
        print(f"File Saved --> {sp}")
```

```python
class_grp = df[df['Amount'] > 0].copy().groupby(['Class'])['Class'].count()
dfgrp = pd.DataFrame(class_grp)
display(dfgrp)

print(f'Percent of Fraud: {round(float((dfgrp.iloc[1] / dfgrp.iloc[0])*100), 2)}%')
```
